# Le principe de multicolinearité

La multicolinéarité est le principe qui dit qu'une variable indépendante aux autres peut être colinéaire avec un ou plusieurs autres variables. Cela a pour conséquence qu'un petit changement dans une variable colinéaire aux autres, peut donner un résultat complètement différent qu'initialement. Pour voir le principe de multicolinéarité plus concrètement, on peut prendre un exemple pratique. Imaginons que nous avons un collègue de travail qui prend avant de travailler toujours un café avec du sucre. Nous voulons alors estimer la productivité de notre collègue avec et sans son café du matin. Ainsi, si nous voulons enlever le café de sa routine, nous enlevons également le sucre des paramètres. Ces deux paramètres dans notre exemple, sucre et café, sont donc colinéaires. Ainsi, lors d'une estimation par moindres carrée, il est possible que des paramètres soient colinéaires entre eux, ce qui peut nuire aux résultats de nos estimations finales.


# Le VIF , facteur d'influence de la variance


Le VIF est le facteur d'influence de la variance et nous sert à avoir un
indice du taux de multicolinéarité dans le traitement de données par
moindre carrée. Estimons que nous avons une équation d'observation telle
que :<br/> $$B = AX + V$$
où B est le vecteur des observations <br/>
     X le vecteur des paramètres <br/>
   A la matrice design <br/>
   V le vecteur des residus
   
   <br/>
   
   
À partir de la matrice design, on peut obtenir la matrice normale N : $$N=A^TPA$$
La matrice de variance covariance , si on considère que tous les paramètres ont le même poids P :
$$Q=N^{-1}$$

Le VIF dans cette situation est donc définie pour le paramètre i par :

 $$VIF_{i}=N_{i,i}Q_{i,i}$$
 
 Le VIF est souvent accompagné d'un équivalent du coefficient de corrélation :
 
 $$cos \theta_i=R_i=\sqrt{1 - \frac{1}{V_i}}$$



<h3>Interpretation du VIF </h3>

<p> On interprète le VIF tel que le minimal soit 1. Un VIF en dessous de 5 est considéré sans multicolinéarité. Un VIF supérieur à 10 est
estimé comme ayant de la multicolinéarité. Ainsi un grand VIF
accompagné d'un petit $\theta$ indique que ce paramètre est un
combinaison linéaire des autres paramètres et est affectée par de la
multicolinéarité.</p>

<h3> Calcul du VIF </h3>

Le VIF (le facteur d'influence de la variance) est un indice de
multicolinéarité que l'on va utiliser pour nos données DORIS.


<p>  Tout d'abord, nous devons transformer les données obtenues par ginx et
locomotiv. À la fin du traitement des données DORIS, on obtient alors
une matrice normale dans un format particulier que l'on doit manipuler
pour obtenir une matrice symétrique carrée.
</p>

```python
import numpy as np
import math
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_matrix():
    f = open("ASC_matrice_normal_jour23490", "r")
    file_header = f.readline().split()
    num_para, num_cond_eq = int(file_header[0])+1, int(file_header[1])
    row = np.array([])
    raw_data = f.readlines()
    normal_matrix = np.zeros((num_para-1,num_para-1))
    read_position = 0
    rows = []
    for i in range(0,num_para):
        row = np.zeros((num_para-1))
        real_row = raw_data[read_position:read_position+i+1]
        for i in range(len(real_row)):
            real_row[i] = float(real_row[i].split()[0])
        read_position += i+1
        rows.append(row)
        row = np.hstack((real_row,row[len(real_row):]))
        normal_matrix[:,i] = row


    for i in range(1,num_para-1):
       normal_transpose = np.transpose(normal_matrix)
       new_part = normal_transpose[i][:i]
       oldPart = normal_matrix[i,i:] 
       new_row = np.hstack((new_part,oldPart))
       normal_matrix[i] = np.hstack((new_part,oldPart))

    f.close()
    return normal_matrix

N=read_matrix()

print(N)
```

Cette matrice correspond à la matrice normale obtenue lors du jour 23490
avec 774 paramètre utilisé lors de l'estimation par moindre carrée.

On considère que lors du traitement des données de DORIS, tous les
paramètres sont de même poids P et que les contraintes appliquées pour
que la matrice N soit inversible sont négligeables dans le calcul du
VIF.

Ainsi, on peut calculer la matrice de variance covariance associée : $$ Q=N^{-1}  $$</p>

Le VIF se calcule alors comme : $$VIF_{i}=N_{i,i}Q_{i,i}$$

```python
def VIF(N):
    Q=np.linalg.inv(N)
    V=[]
    for i in range(len(N)):
        Vi=N[i][i]*Q[i][i]
        V.append(Vi)
    return V

#print(V)
V=VIF(N)
print("\n")
print("La plus grand valeur du VIF est ")
print(max(V))
print("La plus petite valeur du VIF est ")
print(min(V))
print("\n")
```

Pour accompagner le VIF , on peut calcler le coefficient de coefficient de determination $$cos(\theta)^2=R^2$$


```python
def cos_teta(V):
    cos_teta=[]
    for i in range(len(N)):
        if (1/(V[i])>1):
            cos_teta.append("Error")
            print("il y a une erreur")
        else:
            cos_teta.append(math.sqrt(1-(1/(V[i]))))
    return cos_teta

cos_teta=cos_teta(V)
#print(cos_teta)

print("La plus grand valeur du cos teta est ")
print(max(cos_teta))
print("La plus petite valeur du cos teta est ")
print(min(cos_teta))
```
